```python
import random
import matplotlib.pyplot as plt

N = 1000
win1 = 0 #мы не меняли мнение и выиграли
win2 = 0 #мы меняли мнение и выиграли

result1 = []
result2 = []
#меняем мнение
for i in range(N):
  doors = [1,2,3]
  auto = random.choice(doors)
  if auto == 1:
   goats = [2,3]
  elif auto == 2:
    goats = [1,3]
  else:
    goats = [1,2]

  ch = random.choice(doors)#ch-это выбор
  if ch == auto:
    s = random.choice(goats)#s-показываем
  elif ch == goats[0]:
    s = goats[1]
  else:
    s = goats[0]
  doors.remove(ch)
  doors.remove(s)
  if doors[0] == auto:
    win1 += 1
  result1.append(win1)
#не меняем мнение
for i in range(N):
  doors = [1,2,3]
  auto = random.choice(doors)
  if auto == 1:
   goats = [2,3]
  elif auto == 2:
    goats = [1,3]
  else:
    goats = [1,2]

  ch = random.choice(doors)#ch-это выбор
  if ch == auto:
    win2 += 1
  result2.append(win2)

plt.plot(result1, color = 'pink', label= 'меняем мнение')
plt.plot(result2, color = 'purple',  label= 'не меняем мнение')
plt.grid()
plt.xlabel('N')
plt.ylabel('Выигрыш')
plt.legend()
plt.show()

#print("выбрали дверь ", ch)
#print("эту дверь показали ", s)
#print("где автомобиль ", auto)
#print("где козочки ", goats)
```
