Теоретическое решение задачи: P(A)=25/30 * 24/29[ответил на 2-а вопроса 1-го билета]+(25/30 * 5/29 * 24/28)[ответил на 1-ый вопрос 1-го билета и на вопрос из 2-го билета]+(5/30 * 25/29 * 24/28)[ответил на 2-ой вопрос 1-го билета и на вопрос из 2-го билета] Ответ:P(A)=0.935961

```python
import random

N = 100
M = 0
KL = 30
KN = 25

for o in range(0,N):
    attempt = 0 #Количество взятых билетов
    score = 0 #Оценка за экзамен (от 0 до 3 решенных вопросов)

    all_tasks = []
    for i in range(1,31): #Создание списка вопросов
        all_tasks.append(i)

    known_tasks = []
    for i in range(1,26): #Заполнение списка известных вопросов
        x = random.choice(all_tasks)
        while x in known_tasks:
            x = random.choice(all_tasks)
        known_tasks.append(x)

    while attempt <= 3:
        if score < 2:
            task = random.choice(all_tasks)
            if task in known_tasks:
                score += 1
            all_tasks.remove(task);
        attempt += 1
    if score >= 2:
        M += 1

print("N:", N, "Вероятность события:", M/N)
```
