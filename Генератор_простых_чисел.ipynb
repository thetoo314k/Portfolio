```python
class SequenceGenerator:
    def __init__(self, seq_type: str, n: int, low: int = 0, high: int = None):
        self.seq_type = seq_type.lower()
        self.n = n
        self.low = low
        self.high = high

    def gen_primes(self):
        def is_prime(x):
            if x < 2:
                return False
            for i in range(2, int(x ** 0.5) + 1):
                if x % i == 0:
                    return False
            return True

        count, num = 0, max(2, self.low)
        while count < self.n:
            if is_prime(num):
                if self.high is not None and num > self.high:
                    break
                yield num
                count += 1
            num += 1

    def gen_fib(self):
        a, b = 0, 1
        count = 0

        while count < self.n:
            if a >= self.low and (self.high is None or a <= self.high):
                yield a
                count += 1
            a, b = b, a + b

    def generate(self):
        match self.seq_type:
            case "primes" | "простые":
                return self.gen_primes()
            case "fibonacci" | "фибоначчи":
                return self.gen_fib()
            case _:
                raise ValueError("Неизвестный тип последовательности")

try:
    gen_type = input("Введите тип генератора")
    n = int(input("Введите длину последовательности: "))
    low = int(input("Введите нижнюю границу: "))
    high = input("Введите верхнюю границу (Enter — без ограничения): ")

    high = int(high) if high.strip() else None

    generator = SequenceGenerator(gen_type, n, low, high)

    print(f"\nПоследовательность ({gen_type}):")
    for num in generator.generate():
        print(num, end=" ")
    print()

except ValueError as e:
    print("Ошибка:", e)
except Exception as e:
    print("Непредвиденная ошибка:", e)
finally:
    print("\nРабота программы завершена")
```

```python
def run_test(description, seq_type, n, low=0, high=None):
    print(f"{description}")
    print(f"Ввод: type={seq_type}, n={n}, low={low}, high={high}")

    try:
        gen = SequenceGenerator(seq_type, n, low, high)
        result = list(gen.generate())
        print("Вывод:", result)
    except Exception as e:
        print("Ошибка:", e)
    print()

run_test("test 1", "primes", 5)
run_test("test 2", "fibonacci", 7)
run_test("test 3", "простые", 3, 10)
run_test("test 4", "фибоначчи", 5, 1, 10)
run_test("test 5", "abcd", 5)
run_test("test 6", "primes", 0)
```
