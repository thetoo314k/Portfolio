```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([
    [174, 180, 174, 177, 171, 173, 174, 178],
    [175, 177, 174, 176, 171, 168, 176, 177],
    [172, 172, 174, 179, 171, 176, 174, 173],
    [175, 173, 177, 180, 180, 173, 171, 174],
    [176, 171, 173, 177, 173, 167, 169, 175],
    [174, 175, 175, 170, 178, 173, 167, 179]
])

t = np.array([0, 1, 2, 3, 4, 5, 6, 7])

# построение графика
plt.figure(figsize=(10, 6))
for i in range(X.shape[0]):
    plt.plot(t, X[i], label=f'X{i+1}')

# вычисление математического ожидания
mean_X = np.mean(X, axis=1)
print("Математическое ожидание:", mean_X)

# вычисление дисперсии
var_X = np.var(X, axis=1)
print("Дисперсия:", var_X)

std_X = np.std(X, axis=1)
print("стандартное отклонение:", std_X)

# вычисление нормированной корреляционной функции
norm_corr = np.corrcoef(X)
print("Нормированная корреляционная функция:")
print(norm_corr)

plt.xlabel('t')
plt.ylabel('X(t)')

plt.title('График случайной функции X(t)')
plt.legend()
plt.show()
```
