Рассчитать число С(30,7) и время **вычисления**

```python
import time

def comb(n, k):
  if k in (0, n):
    return 1
  elif k!=1:
    return comb(n-1,k) + comb(n-1,k-1)
  else:
   return n

t = time.time() #секундомер
x = comb(30,7)
t = time.time() - t

print(x)
print("Время:", t,)
```

Построить график зависимости временни вычислений от изменения n для С(n, k)

```python
import matplotlib.pyplot as plt
import time

def comb(n, k):
  if k in (0, n):
    return 1
  elif k!= 1:
    return comb(n-1,k) + comb(n-1,k-1)
  else:
   return n

print("Введите количество шагов: ")
N=int(input())
time_list=[]

for i in range(7, N+1):
  t=time.time()
  x=comb(i,7)
  t=time.time() - t
  time_list.append(t)

plt.plot(time_list, label = 'Сочетание', color = 'pink')#строим график
plt.xlabel('Шаги')
plt.ylabel('Время')
plt.legend()
plt.grid()
plt.show()
```
