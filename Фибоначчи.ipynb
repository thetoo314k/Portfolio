```python
# Вычисление чисел Фибоначчи с использованием цикла
def fibonacci_loop(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

# Вычисление чисел Фибоначчи с использованием рекуррентной формулы
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# Рассчитываем число Фибоначчи для n=36
n = 36
print(f"Число Фибоначчи для n={n} (цикл): {fibonacci_loop(n)}")
print(f"Число Фибоначчи для n={n} (рекурсия): {fibonacci_recursive(n)}")
```
