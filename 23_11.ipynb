```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
np.random.seed(4)

dd = 730  # наблюдаемые дни
n_d = 200  # производство булочек в день
wt = 1.0  # вес
unw = 0.05  # недовес
sigma = 0.05  # сигма

x = np.linspace(0.8, 1.2, 100)

# honest очень честный булочник
hon = st.norm.rvs(loc=1.0, scale=sigma, size=dd)

# dishonest нечестный булочник
dis = st.norm.rvs(loc=1.0 - unw, scale=sigma, size=dd)

# Гистограмма относительных частот нечестного булочника
plt.figure(figsize=(10, 6))
plt.hist(dis, bins=30, density=True,label='Для нечестного булочника')
plt.plot(x, st.norm.pdf(x, 1, sigma), linestyle='dashed', color='blue',label='Для честного булочника')
plt.title('Гистограмма относительных частот нечестного булочника')
plt.legend()
plt.show()

# Вычисление выборочного среднего, выборочной дисперсии, коэффициента асимметрии и эксцесса для нечестного булочника
mean_dis = np.mean(dis)
variance_dis = np.var(dis)
skewness_dis = st.skew(dis)
kurtosis_dis = st.kurtosis(dis)

print("Выборочное среднее нечестного булочника:", mean_dis)
print("Выборочная дисперсия нечестного булочника:", variance_dis)
print("Коэффициент асимметрии нечестного булочника:", skewness_dis)
print("Коэффициент эксцесса нечестного булочника:", kurtosis_dis)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

a = 0
b = 2

# Генерируем случайные величины ξ и η
N = 1000
xi = np.random.normal(0, 1, N)
eta = np.random.normal(0, 1, N)

# Вычисляем сумму случайных величин ν
nu = xi**2 + eta**2
plt.hist(nu, bins=20, density=True, alpha=0.5, label='Гистограмма')

#хи квадрат
x = np.linspace(0, 10, 100)
plt.plot(x, chi2.pdf(x, df=2))
plt.xlabel('Значение случайной величины')
plt.ylabel('Эмпирическая функция распределения')
plt.show()
```
