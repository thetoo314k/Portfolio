СР_1 Модель для многоэтапного эксперимента (первый билет, затем, при необходимости, – второй)

```python
import random

N = 100
M = 0
KL = 30
KN = 25

for o in range(0,N):
    attempt = 0 #Количество взятых билетов
    score = 0 #Оценка за экзамен (от 0 до 3 решенных вопросов)

    all_tasks = []
    for i in range(1,31): #Создание списка вопросов
        all_tasks.append(i)

    known_tasks = []
    for i in range(1,26): #Заполнение списка известных вопросов
        x = random.choice(all_tasks)
        while x in known_tasks:
            x = random.choice(all_tasks)
        known_tasks.append(x)

    while attempt <= 3:
        if score < 2:
            task = random.choice(all_tasks)
            if task in known_tasks:
                score += 1
            all_tasks.remove(task);
        attempt += 1
    if score >= 2:
        M += 1

print("N:", N, "Вероятность события:", M/N)
```

```python
import random
import matplotlib.pyplot as plt

N = 1000
win1 = 0 #мы не меняли мнение и выиграли
win2 = 0 #мы меняли мнение и выиграли

result1 = []
result2 = []
#меняем мнение
for i in range(N):
  doors = [1,2,3]
  auto = random.choice(doors)
  if auto == 1:
   goats = [2,3]
  elif auto == 2:
    goats = [1,3]
  else:
    goats = [1,2]

  ch = random.choice(doors)#ch-это выбор
  if ch == auto:
    s = random.choice(goats)#s-показываем
  elif ch == goats[0]:
    s = goats[1]
  else:
    s = goats[0]
  doors.remove(ch)
  doors.remove(s)
  if doors[0] == auto:
    win1 += 1
  result1.append(win1)
#не меняем мнение
for i in range(N):
  doors = [1,2,3]
  auto = random.choice(doors)
  if auto == 1:
   goats = [2,3]
  elif auto == 2:
    goats = [1,3]
  else:
    goats = [1,2]

  ch = random.choice(doors)#ch-это выбор
  if ch == auto:
    win2 += 1
  result2.append(win2)

plt.plot(result1, color = 'pink', label= 'меняем мнение')
plt.plot(result2, color = 'purple',  label= 'не меняем мнение')
plt.grid()
plt.xlabel('N')
plt.ylabel('Выигрыш')
plt.legend()
plt.show()

#print("выбрали дверь ", ch)
#print("эту дверь показали ", s)
#print("где автомобиль ", auto)
#print("где козочки ", goats)
```
